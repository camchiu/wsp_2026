# (1) Fitbit Sleep & Activity Dataset

This dataset is from the Fitabase Fitbit data collection. Each row represents one participant on one day. https://www.kaggle.com/datasets/arashnic/fitbit/data

### participant

    Anonymous identifier for each study participant.

### Day

    Day of the study period, indexed sequentially.

### sleep_hours

    Total sleep duration per night (in hours).

### inbed_hours

    Total time spent in bed per night (in hours), including time awake.

### sleep_efficiency

    Proportion of time in bed spent actually sleeping. 

### Steps

    Total number of steps taken in a day.

### Distance_miles

    Total distance traveled in miles per day.

### Calories_burned

    Total estimated calories burned per day.

### time_sedentary_mins

    Total minutes spent sedentary (inactive) per day.

# (2) BRFSS Sleep and Health Dataset

This dataset is derived from the 2022 Behavioral Risk Factor Surveillance System (BRFSS), a nationwide health survey conducted by the U.S. Centers for Disease Control and Prevention (CDC). Each row represents one adult survey respondent. https://www.cdc.gov/brfss/

### state

    State of residence for adult survey respondent.

### sleep_hours

    Average number of hours of sleep the respondent gets in a 24-hour period.

### general_health

    Respondent's overall rating of their health. (1 = Excellent; 2 = Very Good; 3 = Good; 4 = Fair; 5 = Poor)

### poor_physical_days

    Number of days during the past 30 days when physical health was not good (such as illness or injury).

### poor_mental_days

    Number of days during the past 30 days when mental health was not good (such as stress, depression, or emotional problems).

### exercise

    Whether the respondent participated in any physical activity or exercise during the previous month. (Yes = participated in exercise; No = did not participate in exercise)

### BMI

    Body Mass Index (BMI), calculated from self-reported height and weight (<18.5 = Underweight; 18.5-24.9 = Normal weight; 25-29.9 = Overweight; >30 = Obesity).

### depression

    Whether the respondent has ever been diagnosed with a depressive disorder. (1 = Yes; 2 = No)

### age

    Respondent age in years.

### sex

    Respondent sex.

### income

    Annual household income category (increasing represents increasing household income levels).

### education

    Highest level of education completed (1 = never attended school; 2 = elementary school; 3 = some high school; 4 = high school graduate; 5 = some college/technical school; 6 = college graduate).

### smoking

    Current smoking status (1 = smokes every day; 2 = smokes some days; 3 = does not currently smoke).

### alcohol_use

    Whether respondent has consumed alcohol in the last 30 days.
