Order of operations works in python the same as it does on paper.

Let's think about the following example:
$$ (5-2)^{2}/3 $$

To solve this analytically (by hand), we would do the following:

1.   Solve the expression in parenthesis, i.e. $5-2 = 3$
2.   Square the above; i.e. $3^{2} = 9$
3.   Divide the above by $3$; i.e. $9/3 = 3$.

So our final answer is $3$!  (You may be familiar with the "PEMDAS" acronym; that's what we followed in the steps above.)

## Conditional statements

"If they have my favorite soda at the store, then I'll buy some."

This if/then structure is something we use constantly in our lives -- and it exists in python too!

Breaking down what's happening here: IF (and only if) the first part of our sentence is true (the store has the soda), THEN we will do something (buy the soda).

Python understands the concept of true and false too; these are called "boolean" variables, and you saw them in your Python Primer on Sunday!

````python
True
````

````python
False
````

Let's look at some examples of boolean statements in python:

````python
2 < 10
````

````python
80 > 100
````

````python
4 <= 5
````

````python
6 >= 8
````

````python
2 == 7
````

````python
2 != 7
````

Were any of the results you got surprising?

All boolean statements evaluate to either `True` or `False`.

The cool thing is that we can combine this with the if/then logic and create **conditional statements** where certain lines of code are only executed under specific circumstances! (Just as, in our real-life example, we only buy the soda IF the store has our favorite in stock.)

For example:

````python
a = 10 # assign a value of 10 to the variable a

if a == 10: # check whether the value of the variable a is 10
  print("a is 10")
````

In the above example, the text "a is 10" only printed out *because*, in this case, the expression `a==10` evaluated to `True`.

What if that weren't the case?

````python
a = 7 # assign a value of 7 to the variable a

if a == 10: # check whether the value of the variable a is 10
  print("a is 10")
````

Make sure you understand the difference between these two cases.

Here are another few examples.

Before running each one, see if you can predict what the output will be!

````python
x = 2
y = 3
if x < y:
  print("x is less than y")

if x+y > 10:
  print("the sum of x + y is greater than 10")
````

````python
if True:
  print("Hello!")
````

What if we want to be more flexible though? Think about the logic of the following:

"If they have my favorite soda at the store, then I'll buy some. Otherwise, I'll buy some apple juice."

Here, we've provided a backup plan! We have an alternate course of action if the store doesn't have our favorite soda.

In python we can do the same thing, using the `else` statement. It looks like this:

````python
x = 4
y = 3
if x < y:
  print("x is less than y")
else:
  print("x is not less than y")
````

````python
num_apples = 3
num_bananas = 4
num_grapes = 12

if num_apples + num_bananas > num_grapes:
  print("We need more grapes")
else:
  print("We have enough grapes")
````

## For-loops

For-loops "loop over" the content of lists or arrays. 

For example:

````python
my_list = ["dog", "cat", "bird"]

for value in my_list:
  print(value)
````

Notice the syntax. 

In words, we might say "print each value in this list", or, "for each value in this list, print the value."

In python it's very similar. Look at the first line:


``` python
for value in my_list:
```

Here, `value` is the variable that we use to refer to each individual value within `my_list` as we cycle through. 

After the first line of the for-loop, we have some python code that is indented.  

``` python 
for value in my_list:
  print(value)
```

The indent tells python that `print(value)` needs to be executed *every time* we loop over a new value. This is why we saw this output:

``` python 
dog
cat
bird
```


For-loops can do more than print out values!

Let's look at a couple more examples.

First, we'll use a for-loop to perform mathematical operations:

````python
my_arange = np.array([2, 4, 6, 8])
for value in my_arange:
  doubled = value*2
  print(value, doubled)
````

Now let's try one with conditional statements!

We'll define an array of numbers, and use the for-loop to iterate over them and ask whether or not each one is larger than 10.

````python
my_numbers = np.array([7, -10, 11, 31, 0, -5, 20])

for number in my_numbers:
  print(number > 10)
````

## Conditional indexing

Conditional statements can also help us select **part** of an array.

For example, run the following cell to define an array `my_nums`:

````python
my_nums = np.arange(0,10)
my_nums
````

As we can see, this is an array of numbers with values ranging from 0 to 9.

What if we only want the part of the array where the value is greater than 5?

We can do that as follows:

````python
my_nums[my_nums > 5]
````

The **conditional statement** is enclosed in square brackets, ie `[my_nums > 5]`. This will evaluate to true or false for every element of `my_nums`. 

To see this, run the cell below:

````python
my_nums > 5
````

The first several values are `False`, because these elements correspond to values of 0-5 (and we asked whether they were *greater than* 5).

Think back to for-loops! We're effectively doing an *"in place" for-loop* here by running through each value in `my_nums` and asking whether or not it is larger than 5. The key difference is that rather than explicitly using a for-loop we're just applying a conditional statement to an array.

The **conditional indexing** comes when we put the conditional statement in square brackets and use it to index our array. By doing this, we are asking for *only* the elements where the condition is `True`.

Let's practice this -- in the cell below, index `my_nums` to get all the values less than 4:

Note we can also provide multiple conditional statements. For example, if we want all the values in `my_nums` *between* 2 and 8, we can run the cell below:



````python
my_nums[(my_nums > 2) & (my_nums < 8)]
````

Often we want to be able to visualise something in python, for example an image or a plot. Data visualization is a central part of any exploratory data analysis, and we'll be doing some of it in this lab!

Along the way you can use some pre-written code to make specific plots. However, it's worth going over the basics here first in case you want to make a different kind of plot at some point!

Plotting in python involves heavy usage of a library called `matplotlib`. We can import this in the cell below. 

````python
import matplotlib.pyplot as plt
````

Let's make our first figure.

First, we'll define a variable `x` that's an array running from 0 to 10. Then we'll plot `x` vs `x`.

````python
x = np.arange(10)

plt.plot(x, x, marker = 'o', color = 'black') # plot the data, choose 'o' as the marker symbol and 'black' as the color of the points
plt.xlabel('X') # label the x-axis
plt.ylabel('Y') # label the y-axis
plt.title('My first plot') # give the figure a title
````

What if we want to plot two things on the same figure?

````python
plt.plot(x, x, 'o', color='black') # plot x
plt.plot(x, x**2, 'o', color='orangered') # plot x**2
plt.xlabel('X') # label the x-axis
plt.ylabel('Y') # label the y-axis
plt.title('A more complicated plot') # give the figure a title
````

---

When making figures, we can also be more explicit about creating figure and axes objects. 

You can think of the `Figure` object as a canvas, or a surface to draw on. The `Axes` objects are what you "draw" your points on (and you can place one or more `Axes` object on a single `Figure` object). 

For more details, you can start [with this medium blog](https://medium.com/towards-artificial-intelligence/day-3-of-matplotlib-figure-axes-explained-in-detail-d6e98f7cd4e7). Figures (and matplotlib objects in general) are very well documented online though, so you should be able to find many more resources!

---

In the cell below, we'll define a `Figure` object, and a single `Axes` object, and plot the same data as above:

````python
fig, ax = plt.subplots(1, 1) # 1 row of axes, 1 column of axes (i.e. in this case, 1 set of axes in total)
ax.plot(x, x, 'o', color='black') # we plot on the axes object
ax.plot(x, x**2, 'o', color='orangered')
ax.set_xlabel('X') # notice the slightly different syntax! set_xlabel instead of just xlabel
ax.set_ylabel('Y')
ax.set_title('Axes title'); # give the axes a title
#fig.suptitle('Figure title'); # and/or give the whole figure a title
````

Notice that we are creating the histogram and setting the axis labels directly on `ax`, our `Axes` object (and not `fig`). 

Also, there are slight differences in syntax - for example, where previously we wrote `plt.xlabel()`, now we write `ax.set_xlabel()`.
